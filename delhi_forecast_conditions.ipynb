```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
```

To implement : GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5)
param_grid = {
    'n_estimators': [200,250,300],
    'learning_rate': [0.1, 0.2,0.5],
    'max_depth': [9, 12]
}
xgb = XGBClassifier(random_state=42, n_jobs=-1)
grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='accuracy',
    cv=cv,  # Number of cross-validation folds
    verbose=1,
    n_jobs=-1  # Use all available cores
)
#Fit GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
#Get the best model from gridsearch
best_model = grid_search.best_estimator_
print("Best Parameters:\n", grid_search.best_params_)
print("Best Score:\n", grid_search.best_score_)
# Best Parameters:
#  {'learning_rate': 0.1, 'max_depth': 9, 'n_estimators': 200}
# Best Score:
#  0.7696735521408922
```

```python
from xgboost import XGBClassifier;
model = XGBClassifier(n_estimators=200, learning_rate=0.2,n_jobs = 4,max_depth = 9,random_state = 42)
model.fit(X_train, y_train,
             verbose=False)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
